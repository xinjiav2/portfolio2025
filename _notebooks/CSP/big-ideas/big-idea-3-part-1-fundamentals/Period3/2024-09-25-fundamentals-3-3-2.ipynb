# Add Subtract Multiply and Divide Functions

```python
%%javascript

function add(a, b) {
    return a+b
}

function subtract(a, b) {
    return a-b
}

function multiply(a, b) {
    return a*b
}

function divide(a, b) {
    return a/b
}

function mod(a, b) {
    return a % b
}

function exponent(a, b) {
    return a^b
}
```
